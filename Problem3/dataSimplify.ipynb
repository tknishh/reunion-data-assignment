```python
import sqlite3

# Create a database connection
conn = sqlite3.connect('music.db')
cursor = conn.cursor()

# Create the tables
cursor.execute('''CREATE TABLE Orchestras (
                    orchestra_id INTEGER PRIMARY KEY,
                    name TEXT
                )''')

cursor.execute('''CREATE TABLE Concerts (
                    concert_id INTEGER PRIMARY KEY,
                    orchestra_id INTEGER,
                    date TEXT,
                    venue TEXT,
                    location TEXT,
                    time TEXT,
                    event_type TEXT,
                    FOREIGN KEY (orchestra_id) REFERENCES Orchestras(orchestra_id)
                )''')

cursor.execute('''CREATE TABLE Works (
                    work_id INTEGER PRIMARY KEY,
                    concert_id INTEGER,
                    title TEXT,
                    conductor TEXT,
                    composer_id INTEGER,
                    movement TEXT,
                    FOREIGN KEY (concert_id) REFERENCES Concerts(concert_id),
                    FOREIGN KEY (composer_id) REFERENCES Composers(composer_id)
                )''')

cursor.execute('''CREATE TABLE Soloists (
                    soloist_id INTEGER PRIMARY KEY,
                    work_id INTEGER,
                    name TEXT,
                    roles TEXT,
                    instrument TEXT,
                    FOREIGN KEY (work_id) REFERENCES Works(work_id)
                )''')

cursor.execute('''CREATE TABLE Composers (
                    composer_id INTEGER PRIMARY KEY,
                    name TEXT
                )''')

cursor.execute('''CREATE TABLE Movements (
                    movement_id INTEGER PRIMARY KEY,
                    work_id INTEGER,
                    title TEXT,
                    FOREIGN KEY (work_id) REFERENCES Works(work_id)
                )''')

# Extract and insert data into tables

# Insert Orchestras
orchestras_df = pd.DataFrame(programs['orchestra'], columns=['name'])
orchestras_df['orchestra_id'] = range(1, len(orchestras_df) + 1)
orchestras_df.to_sql('Orchestras', conn, if_exists='append', index=False)

# Insert Concerts
concerts_df = pd.DataFrame(concert[['Date', 'Venue', 'Location', 'Time', 'eventType']], columns=['Date', 'Venue', 'Location', 'Time', 'event_type'])
concerts_df['orchestra_id'] = range(1, len(concerts_df) + 1)
concerts_df.to_sql('Concerts', conn, if_exists='append', index=False)

# Insert Composers
composers_df = pd.DataFrame(work['composerName'], columns=['name'])
composers_df['composer_id'] = range(1, len(composers_df) + 1)
composers_df.to_sql('Composers', conn, if_exists='append', index=False)

# Insert Works
works_df = pd.DataFrame(work[['workTitle', 'conductorName', 'composerName', 'movement']], columns=['title', 'conductor', 'composer_id', 'movement'])
works_df['work_id'] = range(1, len(works_df) + 1)
works_df.to_sql('Works', conn, if_exists='append', index=False)

# Insert Soloists
soloists_df = pd.DataFrame(soloist[['soloistName', 'soloistRoles', 'soloistInstrument']], columns=['name', 'roles', 'instrument'])
soloists_df['soloist_id'] = range(1, len(soloists_df) + 1)
soloists_df.to_sql('Soloists', conn, if_exists='append', index=False)

# Insert Movements
movements_df = pd.DataFrame(movement[['movement']], columns=['title'])
movements_df['movement_id'] = range(1, len(movements_df) + 1)
movements_df.to_sql('Movements', conn, if_exists='append', index=False)

# Commit the changes and close the connection
conn.commit()
conn.close()
```

```python
# convert all dataframes into csv
orchestras_df.to_csv('csv/orchestras.csv', index=False)
concerts_df.to_csv('csv/concerts.csv', index=False)
composers_df.to_csv('csv/composers.csv', index=False)
works_df.to_csv('csv/works.csv', index=False)
soloists_df.to_csv('csv/soloists.csv', index=False)
movements_df.to_csv('csv/movements.csv', index=False)
```
