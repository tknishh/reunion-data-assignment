```python
import sqlite3
import pandas as pd

# Create a connection
conn = sqlite3.connect('orchestra.db')

# Create a cursor
cursor = conn.cursor()

# Create a table
cursor.execute('''
    CREATE TABLE Concerts (
    concert_id INTEGER PRIMARY KEY,
    orchestra_id INTEGER,
    date TEXT,
    venue TEXT,
    location TEXT,
    time TEXT,
    event_type TEXT
    );'''
)

# Insert values
with open('concerts.csv', 'r') as f:
    df = pd.read_csv(f)
    df.to_sql('Concerts', conn, if_exists='append', index=False)

conn.commit()
conn.close()
```

```python
# Create a database connection
conn = sqlite3.connect('orchestra.db')
cursor = conn.cursor()

# Get orders in dataframe
cursor.execute("SELECT * FROM Concerts;")
orders = cursor.fetchall()
concerts_df = pd.DataFrame(orders, columns=[column[0] for column in cursor.description])
```
